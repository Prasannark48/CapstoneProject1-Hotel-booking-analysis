#8.Describe the Meal Preference over customer segment ?

```python
def get_count(series, limit=None): ##let's define a function in a method and initialize the two arguments
  if limit != None:
        series = series.value_counts()[:limit]#count the number of particular meal preffered
  else:
        series = series.value_counts()
  x = series.index
  y = series/series.sum()*100
  return x.values,y.values
```

image.png

```python
##draw the graph of preferred meal by the customers using market segment as a data
plt.figure(figsize = (15,8))
plt.style.use("fivethirtyeight")
meal_count,pct_meal_count = get_count(hotel_data["meal"])
sns.countplot(x = "market_segment",hue = "meal",data = hotel_data)
plt.legend(loc ="upper right")
```

#*Analysis:*

This shows that Bed,Breakfast meal is highly preferred by the customers.Lot of customers bookings done by online TA/TO are preferred 'BB' as a first choice and we can see 'FB' has least choice.
